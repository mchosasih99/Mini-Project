```python
#Drop row that have outliers
df = df.drop([164,  617,  655,  687, 1300, 1653, 2132, 2233])
df.shape
```

```python
#Fill null(on Income) and odd value (on Year_Birth)
df['Income'] = df['Income'].fillna(df['Income'].median())
df['Year_Birth'] = df['Year_Birth'].replace({1893:1970,
                                             1899:1970, 1900:1970})
```

# Extract Columns

```python
#Extract Columns

df['age'] = 2022 - df['Year_Birth']
df['kids'] = df['Kidhome'] + df['Teenhome']
df['total_transactions'] = df['MntCoke'] + df['MntMeatProducts'] + df['MntFruits'] + df['MntFishProducts'] + df['MntSweetProducts'] + df['MntGoldProds']
df['total_purchases'] = df['NumWebPurchases'] + df['NumCatalogPurchases'] + df['NumStorePurchases']
df['Cmpn_Acc'] = df['AcceptedCmp1'] + df['AcceptedCmp2']+ df['AcceptedCmp3'] + df['AcceptedCmp4'] + df['AcceptedCmp5'] 

df.drop(['Kidhome', 'Teenhome', 'Year_Birth'],1,inplace=True)
```

```python
#Age Grouping
conditions = [
    df['age'].isin(list(range(25,35))),
    df['age'].isin(list(range(35,45))),
    df['age'].isin(list(range(45,55))),
    df['age'].isin(list(range(55,65))),
    df['age'].isin(list(range(65,90)))
]

choices = ['young adult', 'middle age adult', 'adult', 'senior adult', 'seniors']

df['age_group'] = np.select(conditions, choices)
df
```

# Extract Membership Columns

```python
df['today'] = pd.to_datetime("today")
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'])
df['member_years'] = ((((df['today'] - df['Dt_Customer']).dt.days).astype(int))/365).round(2)
df.drop(columns=(['today', 'Dt_Customer']), inplace=True)
```

```python
#Categorize consumers based on the length of time they have been members
df['member_category'] = np.where(df.member_years > 9.4, 'loyal',
                                       np.where(df.member_years <8.45, 'new','average'))
df.drop('member_years', axis=1,inplace=True)
```

# Feature Encoding

```python
# Labeling

df_label = df.copy()

label_encoding = {"Education": {'SMA':0,'D3':1,'S1':2, "S2": 3, "S3": 4},
                  "age_group": {'young adult':0, 'middle age adult':1, 'adult':2, 'senior adult':3, 'seniors':4}}

df_label[['Education', 'age_group']]=df_label[['Education', 'age_group']].replace(label_encoding)
```

```python
#One Hot Encoding

for i in ['member_category', 'Marital_Status']:
    onehots = pd.get_dummies(df_label[i], prefix=i)
    df_label = df_label.join(onehots)
    
df_label = df_label.drop(columns=['member_category', 'Marital_Status'], axis =1)
```

# Feature Selection

```python
# Function for deleting redundant features
def cekRedundan(df, treshold):
    df_loop = df.copy()
    redundan = [
                [],[],[]
                ]    
    
    for index, row in df_loop.iterrows():
        for i in row.index.values:
            if (len(redundan[0]) == 0) | ((index+' - '+i not in redundan[2]) & (i+' - '+index not in redundan[2])) :
                nilai = df_loop.loc[index,i]
                if (abs(nilai) >= treshold) & (nilai != 1.0) :
                    redundan[0].append(index)
                    redundan[1].append(i)
                    redundan[2].append(index + ' - ' + i)
                    
    for i in redundan[2]:
        print(i)
```

```python
# Choose redundant features
cekRedundan(df_label.corr(),0.7)
```

```python
# Drop redundant
df_label.drop(['Income','MntCoke','NumCatalogPurchases', 'AcceptedCmp5',
               'MntMeatProducts', 'age', 'Marital_Status_Single',
               'total_purchases'],1,inplace=True)
```
