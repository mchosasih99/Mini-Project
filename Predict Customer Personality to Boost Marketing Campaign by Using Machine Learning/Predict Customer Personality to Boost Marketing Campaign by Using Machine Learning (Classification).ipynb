```python
#Fill null(on Income) and odd value (on Year_Birth)
df['Income'] = df['Income'].fillna(df['Income'].median())
df['Year_Birth'] = df['Year_Birth'].replace({1893:1970, 1899:1970, 1900:1970})
```

## Extract Columns

```python
#Extract Columns

df['age'] = 2022 - df['Year_Birth']
df['kids'] = df['Kidhome'] + df['Teenhome']
df['total_transactions'] = df['MntCoke'] + df['MntMeatProducts'] + df['MntFruits'] + df['MntFishProducts'] + df['MntSweetProducts'] + df['MntGoldProds']
df['total_purchases'] = df['NumWebPurchases'] + df['NumCatalogPurchases'] + df['NumStorePurchases']
df['Cmpn_Acc'] = df['AcceptedCmp1'] + df['AcceptedCmp2']+ df['AcceptedCmp3'] + df['AcceptedCmp4'] + df['AcceptedCmp5'] 

df.drop(['Kidhome', 'Teenhome', 'Year_Birth'],1,inplace=True)
```

```python
#Age Grouping
conditions = [
    df['age'].isin(list(range(25,35))),
    df['age'].isin(list(range(35,45))),
    df['age'].isin(list(range(45,55))),
    df['age'].isin(list(range(55,65))),
    df['age'].isin(list(range(65,90)))
]

choices = ['young adult', 'middle age adult', 'adult', 'senior adult', 'seniors']

df['age_group'] = np.select(conditions, choices)
df
```

## Extract Membership Columns

```python
df['today'] = pd.to_datetime("today")
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'])
df['member_years'] = ((((df['today'] - df['Dt_Customer']).dt.days).astype(int))/365).round(2)
df.drop(columns=(['today', 'Dt_Customer']), inplace=True)
```

## Feature Selection

```python
# Function for deleting redundant features
def cekRedundan(df, treshold):
    df_loop = df.copy()
    redundan = [
                [],[],[]
                ]    
    
    for index, row in df_loop.iterrows():
        for i in row.index.values:
            if (len(redundan[0]) == 0) | ((index+' - '+i not in redundan[2]) & (i+' - '+index not in redundan[2])) :
                nilai = df_loop.loc[index,i]
                if (abs(nilai) >= treshold) & (nilai != 1.0) :
                    redundan[0].append(index)
                    redundan[1].append(i)
                    redundan[2].append(index + ' - ' + i)
                    
    for i in redundan[2]:
        print(i)
```

```python
#Choose redundant features
cekRedundan(df_label.corr(),0.7)
```

```python
# Divide Features and Target

X = df_standard.drop(columns=['Response'], axis=1)
y = df_standard['Response']
```

```python
# Split train and test data

from sklearn.model_selection import train_test_split

#Normal
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,random_state = 42)

#Resampling
from imblearn import under_sampling, over_sampling
X_over, y_over = over_sampling.RandomOverSampler(0.2,random_state = 10).fit_resample(X_train, y_train) #Random Oversampling
X_smote, y_smote = over_sampling.SMOTE(0.2,random_state = 75).fit_resample(X_train, y_train) #SMOTE
```

**The data has imbalanced target features, so resampling is performed to prevent the model from being overfitted. Resampling methods used are Random Oversampling and SMOTE.**

### Hyperparameter Tuning

```python
#Hyperparameter
n_estimators = [int(x) for x in np.linspace(1, 100, num=100)]
max_depth = [int(x) for x in np.linspace(1, 110, num = 30)]
min_samples_split = [int(x) for x in np.linspace(1, 40, num=40)]
min_samples_leaf = [int(x) for x in np.linspace(1, 20, num=20)]
max_features = ['auto', 'sqrt']
criterion = ['gini', 'entropy']
rf_hyperparameters = dict(n_estimators=n_estimators,
                  max_depth=max_depth,
                  min_samples_split=min_samples_split,
                  min_samples_leaf=min_samples_leaf,
                  max_features=max_features,
                  criterion=criterion)
#Initiation
rf = RandomForestClassifier(random_state=42)

#Fit hyperparameter to model
rf1_cv = RandomizedSearchCV(rf, rf_hyperparameters, cv=5, random_state=42, scoring='precision')
rf1_cv.fit(X_train,y_train) #Normal

rf2_cv = RandomizedSearchCV(rf, rf_hyperparameters, cv=5, random_state=42, scoring='precision')
rf2_cv.fit(X_over,y_over) #Random Oversampling

rf3_cv = RandomizedSearchCV(rf, rf_hyperparameters, cv=5, random_state=42, scoring='precision')
rf3_cv.fit(X_smote,y_smote) #SMOTE

#Evaluation
print("Normal:")
eval_classification(rf1_cv)
print("\nRandom Oversampling:")
eval_classification_over(rf2_cv)
print("\nSMOTE:")
eval_classification_smote(rf3_cv)
```

**With hyperparameter tuning, data modelling uses the Random Forest Classifier algorithm, whose best performance is data resampling with the SMOTE method.**
