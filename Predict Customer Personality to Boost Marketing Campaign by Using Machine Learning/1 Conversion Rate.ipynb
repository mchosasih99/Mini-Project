# Conversion Rate

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
pd.set_option('display.max_columns', None)

df = pd.read_csv('https://raw.githubusercontent.com/mchosasih99/dataset/main/marketing_campaign_data.csv')
```

```python
df.info()
```

```python
#Drop unnamed column
df.drop(columns='Unnamed: 0', inplace=True )
```

```python
df.describe()
```

```python
#Visualize columns that could potentially have outliers
sns.boxplot(data=df, x='Income')
```

```python
#Position of the Outlier
print(np.where(df['Income']>150000000))
```

```python
#Drop row that have outliers
df = df.drop([164,  617,  655,  687, 1300, 1653, 2132, 2233])
df.shape
```

```python
#Fill null(on Income) and odd value (on Year_Birth)
df['Income'] = df['Income'].fillna(df['Income'].median())
df['Year_Birth'] = df['Year_Birth'].replace({1893:1970, 1899:1970, 1900:1970})
```

```python
#Extract Columns

df['conversion_rate'] = df['Response']/df['NumWebVisitsMonth']
df['age'] = 2022 - df['Year_Birth']
df['kids'] = df['Kidhome'] + df['Teenhome']
df['total_transactions'] = df['MntCoke'] + df['MntMeatProducts'] + df['MntFruits'] + df['MntFishProducts'] + df['MntSweetProducts'] + df['MntGoldProds']
df['total_purchases'] = df['NumWebPurchases'] + df['NumCatalogPurchases'] + df['NumStorePurchases']
```

```python
#Age Grouping
conditions = [
    df['age'].isin(list(range(25,35))),
    df['age'].isin(list(range(35,45))),
    df['age'].isin(list(range(45,55))),
    df['age'].isin(list(range(55,65))),
    df['age'].isin(list(range(65,90)))
]

choices = ['young adult', 'middle age adult', 'adult', 'senior adult', 'seniors']

df['age_group'] = np.select(conditions, choices)
df
```

```python
# Select columns to analyze conversion rate vs customer demographics

df_cr = df[['ID','Marital_Status','Education','age' ,'age_group', 'kids', 'Income', 'total_purchases', 'total_transactions','conversion_rate']]
df_cr.head()
```

```python
categorical = ['Marital_Status', 'Education', 'age_group']
for i in categorical:
  print(df_cr[i].unique())
```

```python
#Label encoding to see correlation between categorical value and conversion rate
label_encoding = {"Marital_Status": {'Lajang':0, "Bertunangan": 1, "Menikah": 2, "Cerai": 3, "Janda": 4, "Duda": 5},
                  "Education": {'SMA':0,'D3':1,'S1':2, "S2": 3, "S3": 4},
                  "age_group": {'young adult':0, 'middle age adult':1, 'adult':2, 'senior adult':3, 'seniors':4}}

df_cr[['Marital_Status', 'Education', 'age_group']]=df_cr[['Marital_Status', 'Education', 'age_group']].replace(label_encoding)
```

```python
#Conversion Rate Countplot
plt.figure(figsize=(6,6))
sns.countplot(data=df_cr, y='conversion_rate')
plt.title('Conversion Rate Distribution')
plt.show()
```

```python
# Heatmap correlation plot
plt.figure(figsize=(12,12))
sns.heatmap(df_cr.corr(),
            cmap='Greens', annot=True, fmt='.2f')
```

```python
# Conversion Rate pairplot
sns.pairplot(data=df_cr,
             y_vars=['conversion_rate'],
             x_vars=['Marital_Status','Education','age' ,'age_group', 'kids', 'Income', 'total_purchases', 'total_transactions'], palette='Greens')
```

```python
# Conversion Rate scatterplot on features that have correlation
scatter = ['kids', 'Income', 'total_purchases', 'total_transactions']
plt.figure(figsize=(11,10))
for i in range(0, len(scatter)):
    plt.subplot(2, 2, i+1) 
    sns.scatterplot(data=df_cr, y='conversion_rate', x=scatter[i])
    plt.tight_layout()
```
